```python
# Question 1
'''
1.1. Solve for x, y & z where: [05 marks]
Eq1 : x + y + z = 2 | Eq2 : 6x - 4y + 5z = 31 | Eq3: 5x + 2y + 2z = 13 
'''

import numpy as np

# here A is coefficients of x,y and z
A = np.array([[1, 1, 1],
              [6, -4, 5],
              [5, 2, 2]])

# here b is the constants in each equations
b = np.array([2, 31, 13])

# Solve the system of equations
solution = np.linalg.solve(A, b)

print("Solution:", solution)
```

```python
# Question 8.1
'''
8. Group Dataframe
8.1. Group ‘emp_df’ to Create a Table ‘emp_df_gen_edu’ using ‘Gender’ & ‘Education’ having Count of Employees [05 marks]
8.2. Group ‘emp_df’ to Create a Table ‘emp_df_gen_age_sal’ using ‘Gender’ with Average of Variables {Age & Gross_Salary_Lakhs} [05 marks]
'''


data = {
    'Name': ['ABC', 'DEF', 'GHI', 'JKL', 'MNO', 'PQR', 'STU', 'VWX', 'YZA', 'BCD'],
    'Education': ['Graduate', 'Postgraduate', 'Postgraduate', 'Graduate', 'Graduate', 'Postgraduate', 'Graduate', 'Postgraduate', 'Graduate', 'Postgraduate'],
    'Gender': ['Male', 'Female', 'Male', 'Other', 'Female', 'Female', 'Male', 'Male', 'Female', 'Other'],
    'Age': [22, 27, 26, 23, 24, 30, 21, 35, 28, 32],
    'Salary_Lakhs': [6, 15, 20, 5, 10, 18, 12, 14, 7, 8],
    'Bonus%': [12.50, 8.75, 6.25, 10.20, 13.60, 11.40, 9.80, 5.50, 7.75, 14.80],
    'Gross_Salary_Lakhs': [6.7500, 16.3125, 21.2500, 5.5100, 11.3600, 20.0520, 13.1760, 14.7700, 7.5425, 9.1840]
}

emp_df = pd.DataFrame(data)


# Group the DataFrame 'emp_df' by 'Gender' and 'Education' and calculate the count
emp_df_gen_edu = emp_df.groupby(['Gender', 'Education']).size().reset_index(name='Count')

# Print the table 'emp_df_gen_edu'
print(emp_df_gen_edu)


# Question 8.2

# Group the DataFrame 'emp_df' by 'Gender' and calculate the average of 'Age' and 'Gross_Salary_Lakhs'
emp_df_gen_age_sal = emp_df.groupby('Gender')[['Age', 'Gross_Salary_Lakhs']].mean()

# Print the table 'emp_df_gen_age_sal'
print(emp_df_gen_age_sal)


```

```python
#Question 9
'''
9. Create Panel Dataframe from Cross-Sectional Dataframe
9.1. From the following Cross-Sectional Dataframe 'df_cross_section': [05 marks]
Company	2023	2024
ZYX	123	321
WVU	456	654
TSR	789	987
Create the following Panel Dataframe 'df_panel':
Company	Year	Profit
TSR	2023	789
TSR	2024	987
WVU	2023	456
WVU	2024	654
ZYX	2023	123
ZYX	2024	321
'''


import pandas as pd

# Create the Cross-Sectional DataFrame 'df_cross_section'
data = {
    'Company': ['ZYX', 'WVU', 'TSR'],
    '2023': [123, 456, 789],
    '2024': [321, 654, 987]
}
df_cross_section = pd.DataFrame(data)
print(df_cross_section)

# Melt the Cross-Sectional DataFrame to create the Panel DataFrame
df_panel = df_cross_section.melt(id_vars=['Company'], var_name='Year', value_name='Profit')

# Sort the Panel DataFrame by 'Company' and 'Year'
df_panel = df_panel.sort_values(by=['Company', 'Year']).reset_index(drop=True)

# Print the Panel DataFrame 'df_panel'
print(df_panel)
```
